```python
# average die roll with regular throws vs "exploding dice" (when top number is hit (6 on a d6 - a 6 sided die), the die is thrown again and summed to the first throw and can also explode again)
```

```python
import random
import statistics
import matplotlib.pyplot as plt
import numpy as np
```

```python
def die(n, d):
    result = []
    for _ in range(n):
        num = random.randint(1, d)
        result.append(num)
    return result
```

```python
def die_explode(n, d):
    result = []
    for _ in range(n):
        total = 0
        while True:
            num = random.randint(1, d)
            total += num
            if num != d:
                break
        result.append(total)
    return result
```

```python
print(die(10, 4))
print(die_explode(10, 4))
```

```python
print("")
print("d4")
print(statistics.mean(die(1000000, 4)))
print(statistics.mean(die_explode(1000000, 4)))

print("")
print("d6")
print(statistics.mean(die(1000000, 6)))
print(statistics.mean(die_explode(1000000, 6)))

print("")
print("d8")
print(statistics.mean(die(1000000, 8)))
print(statistics.mean(die_explode(1000000, 8)))

print("")
print("d10")
print(statistics.mean(die(1000000, 10)))
print(statistics.mean(die_explode(1000000, 10)))

print("")
print("d12")
print(statistics.mean(die(1000000, 12)))
print(statistics.mean(die_explode(1000000, 12)))

print("")
print("d20")
print(statistics.mean(die(1000000, 20)))
print(statistics.mean(die_explode(1000000, 20)))

```

```python
# Number of throws for each value of d
n = 100  # Adjustable number of throws
d_values = [4, 6, 8]

# Dictionary to store results for each d
results = {}

for d in d_values:
    results[d] = die(n, d)

# Create the scatter plot
plt.figure(figsize=(10, 6))

# Plot each series of throws
for d in d_values:
    throws = results[d]
    x = list(range(1, n + 1))  # X-axis values (throw numbers)
    y = throws  # Y-axis values (die results)
    plt.scatter(x, y, label=f'd = {d}', alpha=0.6)

# Add title and labels
plt.title('Scatter Plot of Die Rolls')
plt.xlabel('Throw Number')
plt.ylabel('Die Result')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Number of throws for each value of d
n_values = [10000, 10000, 10000, 10000, 10000, 10000]  # Adjustable number of throws for each die
d_values = [4, 6, 8, 10, 12, 20]

# Dictionary to store results for each d
results = {}

for d, n in zip(d_values, n_values):
    results[d] = die(n, d)

# Create the scatter plot
plt.figure(figsize=(12, 12))

# First subplot for scatter plot
plt.subplot(2, 1, 1)
for d, n in zip(d_values, n_values):
    throws = results[d]
    x = list(range(1, n + 1))  # X-axis values (throw numbers)
    y = throws  # Y-axis values (die results)
    plt.scatter(x, y, label=f'd = {d}', alpha=0.6)

# Add title and labels
plt.title('Scatter Plot of Die Rolls')
plt.xlabel('Throw Number')
plt.ylabel('Die Result')
plt.legend()
plt.grid(True)

# Second subplot for histogram
plt.subplot(2, 1, 2)
for d in d_values:
    throws = results[d]
    plt.hist(throws, bins=np.arange(1, d + 2) - 0.5, alpha=0.6, label=f'd = {d}', edgecolor='black')

# Add title and labels
plt.title('Histogram of Die Results')
plt.xlabel('Die Result')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
